```python
from pandas import DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = DataFrame([['0.75:1', 0.883, 0.687],
                 ['1:1', 0.901, 0.714],
                 ['1:1.5', 0.918, 0.740],
                 ['1:2.5', 0.932, 0.770]], columns = ['n_rows', 'acc', 'f1'])
b = DataFrame([['0.75:1', 0.746, 0.756],
                 ['1:1', 0.748, 0.746],
                 ['1:1.5', 0.740, 0.718],
                 ['1:2.5', 0.727, 0.679]], columns = ['n_rows', 'acc', 'f1'])
```

```python
plt.figure(figsize=((12,6)))
xs = [i for i, _ in enumerate(a.acc)]
plt.plot(xs, a.acc, 'g-', label='Accuracy')
plt.plot(xs, a.f1, 'r-', label='F1 score')
# plt.axis([0, 4, 0, 1.])
plt.xlabel("number of train samples")
plt.xticks(xs, a.n_rows)
plt.title("Amazon data")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=((12,6)))
xs = [i for i, _ in enumerate(b.acc)]
plt.plot(xs, b.acc, 'g-', label='Accuracy')
plt.plot(xs, b.f1, 'r-', label='F1 score')
# plt.axis([0, 4, .9, 1.])
plt.xlabel("number of train samples")
plt.xticks(xs, b.n_rows)
plt.title("Test data")
plt.legend()
plt.show()
```
