```python
from pandas import DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = DataFrame([['100K', 0.938, 0.785],
                 ['200K', 0.939, 0.709],
                 ['500K', 0.939, 0.747],
                 ['1000K', 0.940, 0.760],
                 ['all', 0.941, 0.763]], columns = ['n_rows', 'acc', 'f1'])
```

```python
b = DataFrame([['100K', 0.617, 0.420],
                 ['200K', 0.649, 0.396],
                 ['500K', 0.650, 0.490],
                 ['1000K', 0.667, 0.546],
                 ['all', 0.675, 0.555]], columns = ['n_rows', 'acc', 'f1'])
```

```python
plt.figure(figsize=((12,6)))
xs = [i for i, _ in enumerate(a.acc)]
plt.plot(xs, a.acc, 'g-', label='Accuracy')
plt.plot(xs, a.f1, 'r-', label='F1 score_neg')
# plt.axis([0, 4, 0, 1.])
plt.xlabel("number of train samples")
plt.xticks(xs, a.n_rows)
plt.title("Amazon data")
plt.ylim([0.7, 1])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=((12,6)))
xs = [i for i, _ in enumerate(b.acc)]
plt.plot(xs, b.acc, 'g-', label='Accuracy')
plt.plot(xs, b.f1, 'r-', label='F1 score_neg')
# plt.axis([0, 4, .9, 1.])
plt.xlabel("number of train samples")
plt.xticks(xs, b.n_rows)
plt.title("Test data")
plt.legend()
plt.show()
```
